```python
import re

suffix_map = {
    'STREET': 'St',
    'AVENUE': 'Ave',
    'ROAD': 'Rd',
    'BOULEVARD': 'Blvd',
    'DRIVE': 'Dr',
    'COURT': 'Ct',
    'TERRACE': 'Ter',
    'PLACE': 'Pl',
    'LANE': 'Ln'
}
```

```python
def abbreviate_street_names(name):
    for full_suffix, abbr_suffix in suffix_map.items():
        pattern = re.compile(r'\b' + re.escape(full_suffix) + r'\b', re.IGNORECASE)
        if pattern.search(name):
            name = pattern.sub(abbr_suffix, name)
            break
    return name

# Apply the function to the DataFrame
filterdf['fullStreetName'] = filterdf['fullStreetName'].apply(abbreviate_street_names)
```

```python
filterdf
```

```python
directions = ['North', 'N', 'South', 'S', 'East', 'E', 'West', 'W']
```

```python
def remove_cardinal_directions(name):
    # Create a regex pattern to match any of the directions followed by optional whitespace
    pattern = r'\b(' + '|'.join(re.escape(dir) for dir in directions) + r')\b\s*'
    # Replace occurrences with an empty string
    return re.sub(pattern, '', name, flags=re.IGNORECASE).strip()

filterdf['fullStreetName'] = filterdf['fullStreetName'].apply(remove_cardinal_directions)
filterdf
```

```python
import os
import glob
import pandas as pd
from datetime import datetime, timedelta

# Define the path to the directory containing the CSV files
path = r'C:\Users\marke\projects\parking\ParkingData\scr\CollectedData'  # Use your path

# Get a list of all CSV files in the directory
all_files = glob.glob(os.path.join(path, "*.csv"))

print(all_files)  # Print the list of files to verify

# Process each CSV file
for file in all_files:
    # Read the CSV file
    df = pd.read_csv(file, index_col=None, header=0)
    
    # Add 'Type #' column by extracting the first part of the 'Type' column
    df['Type #'] = df['Type'].str.split(" ").str[0]
    
    # Convert 'Date Issue' column to datetime
    df['Date Issue'] = pd.to_datetime(df['Date Issue'])
    
    # Add 'Hour' and 'DayOfWeek' columns
    df['Hour'] = df['Date Issue'].dt.hour
    df['DayOfWeek'] = df['Date Issue'].dt.day_name()

    # Save the processed data back to the CSV file
    df.to_csv(file, index=False)
```
